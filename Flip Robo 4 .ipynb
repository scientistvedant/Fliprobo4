## Scrap the details team India's international fixtures from bcci.tv.

```python
# creating the driver 
driver = webdriver.Chrome('chromedriver2.exe')

# accessing the website
driver.get('https://www.bcci.tv/')
time.sleep(1)

# selecting the drop down
fixtures = driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]').click()

time.sleep(1)

# Accessing the International Fixtures
fixtures_selection = driver.find_element_by_xpath('/html/body/div[3]/div/div[2]/div[2]/nav/ul/li[1]/div[2]/div/ul/li[1]/a').click()
time.sleep(1)

#creating the empty dataset
Match_title = []
Series_1 = []
Series = []
Place = []
Date = []
Time = []

#fetching data of match title
for i in driver.find_elements_by_xpath('//p[@class="fixture__additional-info"]//strong'):
        Match_title.append(i.text)

#fetching details of type of Series
for n in driver.find_elements_by_xpath('//div[@class="fixture__format-strip"]//span'):
    Series_1.append(n.text)
for i in range(0,len(Series_1)-1,2):
    Series.append(Series_1[i])

#fetching details of venue
for p in driver.find_elements_by_xpath("//p[@class='fixture__additional-info']//span"):
    Place.append(p.text)

#fetching details of Date
for d in driver.find_elements_by_xpath("//span[@class='fixture__date']"):
    Date.append(d.text)

#fetching details of time of play
for t in driver.find_elements_by_xpath("//span[@class='fixture__time']"):
    Time.append(t.text)

#creating the dataframe
International_Cricket = pd.DataFrame({})
International_Cricket['Match_Title'] = Match_title
International_Cricket['Series'] = Series
International_Cricket['Place'] = Place
International_Cricket['Date'] = Date
International_Cricket['Time'] = Time

International_Cricket
```

## Scrap the details of selenium exception from guru99.com

```python
# creating the driver 
driver = webdriver.Chrome('chromedriver2.exe')

#accessing the website
driver.get('https://www.guru99.com/')
time.sleep(1)

#accessing the right choice i.e. selenium topic
selective = driver.find_element_by_xpath("//a[@title='Selenium']").click()
time.sleep(1)

#clicking on the exception list
selective_exception = driver.find_element_by_xpath("//a[@title='Selenium Exception Handling (Common Exceptions List)']").click()
time.sleep(1)

#creating the empty database
exception = []
exception_name = []
description = []

#fetching details of exceptions & descriptions
for i in driver.find_elements_by_xpath("//table[@class='table table-striped']//td"):
    exception.append(i.text)
    
# destributing of seperating the exceptions and Descriptions
for i in range(0,len(exception)-1,2):
    exception_name.append(exception[i])
    description.append(exception[i+1])

#creating the dataframe for the output
Selenium_Exception = pd.DataFrame({})
Selenium_Exception['Exception'] = exception_name
Selenium_Exception['Description'] = description

Selenium_Exception
```

## Scrap the details of State-wise GDP of India from statistictime.com

```python
# creating the driver 
driver = webdriver.Chrome('chromedriver2.exe')

#accessing the website
driver.get('https://statisticstimes.com/')
time.sleep(1)

#clicking on the Economy dropdown
dropdown = driver.find_element_by_xpath('//*[@id="top"]/div[2]/div[2]/button').click()

#clicking on the country i.e. India
select = driver.find_element_by_xpath('//*[@id="top"]/div[2]/div[2]/div/a[3]').click()
```

```python
#selecting the GDP Statewise option
stateGDP = driver.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/ul/li[1]/a').click()

#creating the dummy databases
Rank = []
State = []
GSDP = []
GSDPCP = []
GSDP1 = []
share = []
Share = []
Gdp = []
GDP = []

#feting the details of rank
for i in driver.find_elements_by_xpath('//td[@class="data1"]'):
    Rank.append(i.text)
    
#fetchin the details of State
for s in driver.find_elements_by_xpath('//td[@class="name"]'):
    State.append(s.text)
    
#fetching the details of GSDP (19-20)
for g in driver.find_elements_by_xpath('//td[@class="data"]'):
    GSDP.append(g.text)
for h in range(0,len(GSDP)-1,5):
    GSDPCP.append(GSDP[h])

#fetching the details of GSDP (18-19)
for j in driver.find_elements_by_xpath('//td[@class="data sorting_1"]'):
    GSDP1.append(j.text)

#fetching the details of Share
for k in driver.find_elements_by_xpath('//td[@class="data"]'):
    share.append(k.text)
for l in range(0,len(share)-1,5):
    Share.append(share[l+1])
    
#fetching the details of GDP
for m in driver.find_elements_by_xpath('//td[@class="data"]'):
    Gdp.append(m.text)
for n in range(0,len(Gdp)-1,5):
    GDP.append(Gdp[n+2])
    
#creating the dataframe
GDP_India = pd.DataFrame({})
GDP_India['Rank'] = Rank[:33]
GDP_India['State'] = State[:33]
GDP_India['GSDP1'] = GSDPCP[:33]
GDP_India['GSDP2'] = GSDP1[:33]
GDP_India['Share'] = Share[:33]
GDP_India['GDP($billion)'] = GDP[:33]

GDP_India
```

## Scrap the details of top 100 songs on billboard.com

```python
# creating the driver 
driver = webdriver.Chrome('chromedriver2.exe')

#accessing the website
driver.get('https://www.billboard.com/')

#initiating the chart option
chart = driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/nav/ul/li[6]/a').click()

#clicking on the 100 hot songs
hot_songs = driver.find_element_by_xpath('//*[@id="root"]/div[2]/div[2]/div/div[3]/div[1]/div[2]/div/div[2]/a[1]/div').click()
```

```python
#creating the empty databases
Song_name = []
Artist_name = []
Last_week_Rank = []
Peak_Rank = []
Weeks_on_Board = []

#fetching the details of Songname
for s in driver.find_elements_by_xpath('//span[@class="chart-element__information__song text--truncate color--primary"]'):
    Song_name.append(s.text)
    
#fetching the details of Artistname
for a in driver.find_elements_by_xpath('//span[@class="chart-element__information__artist text--truncate color--secondary"]'):
    Artist_name.append(a.text)

#fetching the details of Last week Rank
for l in driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--last"]'):
    Last_week_Rank.append(l.text)

#fetching the details of Peak Rank
for p in driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--peak"]'):
    Peak_Rank.append(p.text)

#fetching the details of Weeks_on_Board
for w in driver.find_elements_by_xpath('//div[@class="chart-element__meta text--center color--secondary text--week"]'):
    Weeks_on_Board.append(w.text)
    

#creating the dataframe
Top_Songs = pd.DataFrame({})
Top_Songs['Song_Name'] = Song_name
Top_Songs['Artist_Name'] = Artist_name
Top_Songs['Last_week_Rank'] = Last_week_Rank
Top_Songs['Peak_Rank'] = Peak_Rank
Top_Songs['Weeks_on_Board'] = Weeks_on_Board

Top_Songs
```

## scrap the details of Highest Selling Novels

```python
# creating the driver 
driver = webdriver.Chrome('chromedriver2.exe')

#accessing the website
driver.get('https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare')
```

```python
#creating the empty datasets
Book = []
Book_Name = []
Author = []
Sales = []
Publisher = []
Genre = []

#fetching details of Book Name
for b in driver.find_elements_by_xpath("//td[@class='left']"):
    Book.append(b.text)
for c in range(0,len(Book)-1,5):
    Book_Name.append(Book[c+1])

#fetching details of Author Name
for d in range(0,len(Book)-1,5):
    Author.append(Book[d+2])
    
#fetching details of Sales
for s in range(0,len(Book)-1,5):
    Sales.append(Book[s+3])
    
#fetching details of Publisher
for p in range(0,len(Book)-1,5):
    Publisher.append(Book[p+4])

#fetching details of Genre
for g in driver.find_elements_by_xpath("//td[@class='last left']"):
    Genre.append(g.text)
    
#creating the dataframe
High_Selling_Novels = pd.DataFrame({})
High_Selling_Novels['Book_Name'] = Book_Name
High_Selling_Novels['Author_Name'] = Author
High_Selling_Novels['Volumes_Sold'] = Sales
High_Selling_Novels['Publisher'] = Publisher
High_Selling_Novels['Genre'] = Genre

High_Selling_Novels
```
